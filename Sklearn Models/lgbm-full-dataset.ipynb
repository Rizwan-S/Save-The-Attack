# MODEL 1

```python
df_subset = pd.read_csv('../input/trainandtestnew/train1.csv', dtype = dtypes)
import re
df_subset = df_subset.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))
```

```python
main_X = df_subset.drop(['HasDetections'], axis=1)
main_y = df_subset['HasDetections']
```

```python
params = {'max_depth': -1,
          'num_estimators': 30000,
          'metric': 'auc',
          'learning_rate': 0.1,
          'colsample_bytree': 0.2,
          'objective': 'binary', 
          'n_jobs': -1
         }

data_matrix = lgb.Dataset(data=main_X, label=main_y, free_raw_data=False)
```

```python
model1 = lgb.train(params, data_matrix, num_boost_round = 100)
model1.save_model('model1.model')
```

# Model 2

```python
df_subset = pd.read_csv('../input/trainandtestnew/train2.csv', dtype = dtypes)
df_subset = df_subset.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))
```

```python
main_X = df_subset.drop(['HasDetections'], axis=1)
main_y = df_subset['HasDetections']
data_matrix = lgb.Dataset(data=main_X, label=main_y, free_raw_data=False)
```

```python
model2 = lgb.train(params, data_matrix,num_boost_round = 100, init_model='model1.model')
model2.save_model('model2.model')
```

# Model 3

```python
df_subset = pd.read_csv('../input/trainandtestnew/train3.csv', dtype = dtypes)
df_subset = df_subset.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))
```

```python
main_X = df_subset.drop(['HasDetections'], axis=1)
main_y = df_subset['HasDetections']
data_matrix = lgb.Dataset(data=main_X, label=main_y, free_raw_data=False)
```

```python
model3 = lgb.train(params, data_matrix, num_boost_round = 100, init_model='model2.model')
```

```python
test_df = pd.read_csv('../input/trainandtestnew/test_df.csv', dtype = dtypes)
test_df = test_df.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))
```

```python
kaggle_test_X = test_df.drop(columns="MachineIdentifier")
kaggle_test_yhat = model3.predict(kaggle_test_X)
test_df["HasDetections"] = kaggle_test_yhat
to_submit = test_df[["MachineIdentifier", "HasDetections"]]
to_submit.to_csv("class-lgb.csv", index=False)
```
